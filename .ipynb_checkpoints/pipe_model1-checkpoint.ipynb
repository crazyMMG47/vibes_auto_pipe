```python
import scipy.io
import numpy as np
import os 
import matplotlib.pyplot as plt
```

```python
# Load the matlab file
data = scipy.io.loadmat('/home/ecbm4040/vibes_auto_pipe/final_slices.mat')
print(type(data))
x = data.keys()
print(x)
```

```python
final_slices = data["final_slices"]
```

```python
print(final_slices.shape)
```

```python
# Visualize the final slices we want to make prediction here
if final_slices.shape == (160, 160, 48):
    final_slices = np.transpose(final_slices, (2, 0, 1))  # Shape: (48, 160, 160)
final_slices = np.expand_dims(final_slices, axis=-1)  # Shape: (48, 160, 160, 1)
print(final_slices.shape)
```

```python
num_slices_to_show = 48
for i in range(num_slices_to_show):
    plt.figure()
    plt.imshow(final_slices[i, :, :, 0], cmap='gray')  # Show the first grayscale channel
    plt.title(f"Slice {i + 1}")
    plt.axis('off')  # Remove axis
    plt.show()
```

```python
rotated_image = np.array([np.rot90(slice, k=1) for slice in final_slices])
# rotated_image = np.expand_dims(rotated_image, axis=-1)  # Add channel dimension
print(rotated_image.shape)  # Should output: (48, 160, 160, 1)
```

```python
# feed the final_slices into the NN and make predictions
from tensorflow.keras.models import load_model
from loss_funcs import loss_function

# create the instance
my_loss_func = loss_function()

dice_coefficient = my_loss_func.dice_coefficient
dice_loss = my_loss_func.dice_loss

```

```python
saved_model_path = "/home/ecbm4040/vibes_auto_pipe/unet_w_aug.h5"
model = load_model(saved_model_path, custom_objects={"dice_loss": dice_loss, "dice_coefficient": dice_coefficient})
print("Model Loaded Successfully!")


```

```python
print(rotated_image.shape)
```

```python
print(len(rotated_image))
print("Original rotated_image shape:", rotated_image.shape)
print("Final slices shape:", final_slices.shape)
```

```python
# Run predictions on the rotated image and generate the output into a pdf first
# Binarize predictions with a threshold
predictions = model.predict(rotated_image)  # Output shape: (48, 160, 160, 1)
print(predictions.shape)

binary_predictions = (predictions > 0.5).astype(int)

# Optionally squeeze the channel dimension for easier visualization
binary_predictions = binary_predictions.squeeze()  # Shape: (48, 160, 160)
print(binary_predictions.shape)
1
import matplotlib.pyplot as plt

num_slices_to_show = 48 # Number of slices to display
for i in range(num_slices_to_show):
    plt.figure(figsize=(10, 5))

    # Input image
    plt.subplot(1, 2, 1)
    plt.title("Input Image")
    plt.imshow(rotated_image[i, :, :, 0], cmap='gray')
    plt.axis('off')

    # Predicted mask
    plt.subplot(1, 2, 2)
    plt.title("Predicted Mask")
    plt.imshow(binary_predictions[i], cmap='gray')
    plt.axis('off')

    plt.show()
```
